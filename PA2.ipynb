## Task 3: Working with Running Times

**Question:**  
Create two variables representing running times (in minutes).  
Using these variables, calculate and print:
1. The total time spent running  
2. The difference between the two runs  
3. The time for doing the second run twice  
4. The average running time


```python
run1 = 30
run2 = 25


total_time = run1 + run2
print("Total time spent running:", total_time, "minutes")
improvement = run1 - run2
print("Difference between run1 and run2:", improvement, "minutes")
double_run2 = run2 * 2
print("Time for doing run2 twice:", double_run2, "minutes")
average_time = (run1 + run2) / 2
print("Average running time:", average_time, "minutes")
```

## Task 5: User Input and Operations

**Question:**  
Use the Pythagorean theorem to compute the hypotenuse of a right triangle.  

Steps:  
1. Prompt the user to enter the lengths of the two legs (a and b).  
2. Compute the hypotenuse using the formula: c = √(a² + b²)  
3. Print the hypotenuse rounded to the nearest hundredth in the format:  
   “The hypotenuse of your triangle is <c> units.”


```python
import math  


a = float(input("Enter the length of the first leg (a): "))
b = float(input("Enter the length of the second leg (b): "))

c = math.sqrt(a**2 + b**2)


print(f"The hypotenuse of your triangle is {round(c, 2)} units.")
```
