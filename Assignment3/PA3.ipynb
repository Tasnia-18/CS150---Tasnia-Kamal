## Task 1: Types of Triangles (10 pts)
Create a program that takes three side lengths as user input and determines whether 
the triangle is **equilateral**, **isosceles (but not equilateral)**, or **scalene**.

```python
side1 = float(input("Enter side 1: "))
side2 = float(input("Enter side 2: "))
side3 = float(input("Enter side 3: "))

if side1 == side2 == side3:
    print("This is an equilateral triangle.")
elif side1 == side2 or side2 == side3 or side1 == side3:
    print("This is an isosceles (but not equilateral) triangle.")
else:
    print("This is a scalene triangle.")
```

## Task 2: Student Grades and At-Risk Status (12 pts)

For this task you are to do the following:

A. Read five input grade averages (which should be floats) representing student grades 
at the mid-term break into an initially empty list called *student_grades*.

B. First calculate and print the class average, minimum grade, and maximum grade.

C. Any student with a grade below 65 is considered "at risk" -- as such, you should 
print a message indicating how many of the students in the course are at risk. 
Ex: if the student grades are [80, 90, 61, 52, 77], the message should be 
*"2 students are at risk in the course."*

```python

student_grades = []

for i in range(5):
    grade = float(input("Enter grade for student " + str(i + 1) + ": "))
    student_grades.append(grade)


class_average = sum(student_grades) / 5
print("Class Average:", round(class_average, 2))
print("Minimum Grade:", min(student_grades))
print("Maximum Grade:", max(student_grades))


at_risk_count = 0
for grade in student_grades:
    if grade < 65:
        at_risk_count += 1

print(at_risk_count, "students are at risk in the course.")
```

## Task 3: Ranges and Multiples (13 pts)

A. Create a list called *mult_of_2* that contains all multiples of 2 between 1 and 100.

B. Create a list called *mult_of_3* that contains all multiples of 3 between 1 and 100.

C. Create a list called *mult_of_both* that contains all elements that are present in 
**both** mult_of_2 *and* mult_of_3.

D. Create a fourth list called *mult_of_either* that contains all numbers present in 
**either** mult_of_2 *or* mult_of_3.


```python
mult_of_2 = []
for i in range(1, 101):
    if i % 2 == 0:
        mult_of_2.append(i)


mult_of_3 = []
for i in range(1, 101):
    if i % 3 == 0:
        mult_of_3.append(i)


mult_of_both = []
for i in mult_of_2:
    if i in mult_of_3:
        mult_of_both.append(i)


mult_of_either = []
for i in range(1, 101):
    if i % 2 == 0 or i % 3 == 0:
        mult_of_either.append(i)

print("Multiples of 2:", mult_of_2)
print("Multiples of 3:", mult_of_3)
print("Multiples of both:", mult_of_both)
print("Multiples of either:", mult_of_either)
```


## Task 4: Shopping List (15 pts)

A. Starting with an initially empty list named *shopping_list,* allow the user to 
input **up to** 7 items they want to purchase (an input value like "done" should 
be used to stop input early.)

B. Print the total number of items on your shopping list.

C. Print the list in reverse order (opposite order from which the values were input).

D. Print the list in alphabetical order.

E. If any item appears **at least twice** in the list, let the person know as

```python
shopping_list = []

for i in range(7):
    item = input("Enter item " + str(i + 1) + " (or 'done' to stop): ")
    if item.lower() == "done":
        break
    shopping_list.append(item)


print("Total number of items:", len(shopping_list))

# C. Print the list in reverse order
reverse_list = shopping_list[::-1]
print("Shopping list in reverse order:", reverse_list)


sorted_list = sorted(shopping_list)
print("Shopping list in alphabetical order:", sorted_list)


for i in range(len(shopping_list)):
    for j in range(i + 1, len(shopping_list)):
        if shopping_list[i] == shopping_list[j]:
            print(shopping_list[i], "appears at least twice in the list.")
            break
```
